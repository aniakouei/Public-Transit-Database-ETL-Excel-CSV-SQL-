```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.sql import text
from datetime import date, datetime

DB_NAME="publictransportation"
DB_USER="postgres"
DB_PASS="hellosql"
DB_HOST="localhost"
DB_PORT="5432"
```

```python
engine=create_engine(f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}/{DB_NAME}")
conn=engine.connect()

sql_query= """
CREATE TABLE if not exists monthly_totals
(id serial,
month varchar,
year int,
city varchar,
transportation_type varchar,
ridership_count int
);
"""

conn.execute(text(sql_query))
```

```python
conn.commit()
```

```python
nyc_df = pd.read_csv('MTA_Monthly_Ridership___Traffic_Data__Beginning_January_2008.csv')
chicago_df = pd.read_csv('CTA_-_Ridership_-_Daily_Boarding_Totals_20240307.csv')
```

```python
chicago_df['service_date'] = pd.to_datetime(chicago_df['service_date'])
chicago_df.head()
```

```python
nyc_df['Month'] = pd.to_datetime(nyc_df['Month'])
nyc_df.head()
```

```python
chicago_df=chicago_df.drop(columns=['day_type', 'total_rides'])
chicago_df.head()
```

```python
new_chicago_df=pd.melt(chicago_df, id_vars=['service_date'], value_vars=['bus', 'rail_boardings'], var_name='transportation_type', value_name='ridership_count')
new_chicago_df.head()
```

```python
new_chicago_df['transportation_type'].mask(new_chicago_df['transportation_type']== 'rail_boardings', 'metro', inplace=True)
```

```python
new_chicago_df['city']='Chicago'
new_chicago_df.head()
```

```python
def find_month(date):
    if date.month == 1:
        return 'January'
    elif date.month == 2:
        return 'February'
    elif date.month == 3:
        return 'March'
    elif date.month == 4:
        return 'April'
    elif date.month == 5:
        return 'May'
    elif date.month == 6:
        return 'June'
    elif date.month == 7:
        return 'July'
    elif date.month == 8:
        return 'August'
    elif date.month == 9:
        return 'September'
    elif date.month == 10:
        return 'October'
    elif date.month == 11:
        return 'November'
    elif date.month == 12:
        return 'December'
    else:
        return 'Invalid month'
    
new_chicago_df['month']=new_chicago_df['service_date'].apply(find_month)
new_chicago_df.head()
```

```python
new_chicago_df['year'] = new_chicago_df['service_date'].apply(lambda date: date.year)
new_chicago_df.head()
```

```python
final_chicago_df=new_chicago_df.drop(columns=['service_date'])
final_chicago_df=final_chicago_df.groupby(['year','month','transportation_type','city'])['ridership_count'].sum().reset_index()
final_chicago_df.head()
```

```python
nyc_df['city']='New York'
nyc_df.head()
```

```python
nyc_df['month']=nyc_df['Month'].apply(find_month)
nyc_df.head()
```

```python
nyc_df['year'] = nyc_df['Month'].apply(lambda date: date.year)
nyc_df.head()
```

```python
new_nyc_df=nyc_df.rename(columns={"Ridership":"ridership_count"})
```

```python
new_nyc_df['Agency'].value_counts()
```

```python
def transport_type(agency):
    if agency=="MNR" or agency=="Subway"or agency=="LIRR" or agency=="SIR":
        return "metro"
    else:
        return "bus"
    

new_nyc_df=new_nyc_df[new_nyc_df['Agency']!= "B&T"]
new_nyc_df['Agency'].value_counts()
```

```python
new_nyc_df['transportation_type']=new_nyc_df['Agency'].apply(transport_type)
```

```python
display(new_nyc_df)
```

```python
final_nyc_df = new_nyc_df.groupby(['transportation_type', 'year', 'month'])['ridership_count'].sum().reset_index()
display(final_nyc_df)
```

```python
combined_df = pd.concat([final_chicago_df, final_nyc_df])
display(combined_df)
```

```python
for index, row in combined_df.iterrows():
    values= tuple(row)
    query=text(f"insert into monthly_totals (year, month, transportation_type, city, ridership_count) values {values}")
    conn.execute(query)
```

```python
combined_df.to_sql(name='monthly_totals', con=conn, index=False)
```
